# **LOGISTIC REGRESSION** - TITANIC DATA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = sns.load_dataset('titanic')
df
```

```python
df = df.select_dtypes(exclude=int)
df
```

FOR NOMINAL DATA -> LabelEncoder

FOR ORDINAL DATA -> OrdinalEncoder

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['sex'] = le.fit_transform(df[['sex']])
df['embark_town'] = le.fit_transform(df[['embark_town']])
df['who'] = le.fit_transform(df[['who']])
df['alive'] = le.fit_transform(df[['alive']])
df
```

```python
df['who'].unique()
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder(categories = [['First', 'Second', 'Third']])
df['class'] = oe.fit_transform(df[['class']])
df
```

```python
df['class'].unique()
```

```python
df.drop(columns=['embarked'], inplace = True)
df.drop(columns = ['adult_male'], inplace = True)
df.drop(columns = ['deck'], inplace = True)
df.drop(columns = ['alone'], inplace = True)
df.drop(columns = ['who'], inplace = True)
df
```

```python
df.isnull().sum()
```

```python
df['age'].fillna(df['age'].median(), inplace=True)
df.isnull().sum()
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, train_size = 0.8)
model.fit(x_train, y_train)
```

```python
x
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score, recall_score
accuracy_score(y_test, y_pred)
```

```python
recall_score(y_test, y_pred)
```
